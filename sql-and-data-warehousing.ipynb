# DATA WAREHOUSING

```sql
CREATE DATABASE ORDERDWH;
```

### STAGING LAYER

```sql
CREATE OR REPLACE TABLE ORDERDWH.STG_SALES
AS
SELECT * FROM SALES_NEW.ORDERS;
```

### TRANSFORMATION

```sql
CREATE VIEW ORDERDWH.TRANS_SALES
AS
SELECT * FROM ORDERDWH.STG_SALES
WHERE QUANTITY IS NOT NULL;
```

```sql
SELECT * FROM ORDERDWH.TRANS_SALES;
```

### CORE LAYER

##### DimCustomers

```sql
CREATE OR REPLACE TABLE ORDERDWH.DIMCUSTOMERS
(
  DIMCUSTOMERKEY INT,
  CUSTOMERID INT,
  CUSTOMERNAME STRING,
  CUSTOMEREMAIL STRING
);
```

```sql
CREATE OR REPLACE VIEW ORDERDWH.VIEW_DIMCUSTOMERS AS
SELECT 
  ROW_NUMBER() OVER (ORDER BY CUSTOMERID) AS DIMCUSTOMERKEY,
  CUSTOMERID,
  CUSTOMERNAME,
  CUSTOMEREMAIL
FROM (
  SELECT DISTINCT 
    CUSTOMERID,
    CUSTOMERNAME,
    CUSTOMEREMAIL
  FROM ORDERDWH.TRANS_SALES
) AS T;
```

```sql
INSERT INTO ORDERDWH.DIMCUSTOMERS
SELECT * FROM ORDERDWH.VIEW_DIMCUSTOMERS;
```

```sql
SELECT * FROM ORDERDWH.DIMCUSTOMERS;
```

##### DIMPRODUCTS

```sql
CREATE TABLE ORDERDWH.DIMPRODUCTS
(
  DIMPRODUCTKEY INT,
  PRODUCTID INT,
  PRODUCTNAME STRING,
  PRODUCTCATEGORY STRING
)
```

```sql
CREATE OR REPLACE VIEW ORDERDWH.VIEW_DIMPRODUCTS AS
SELECT 
  ROW_NUMBER() OVER (ORDER BY PRODUCTID) AS DIMPRODUCTKEY,
  PRODUCTID,
  PRODUCTNAME,
  PRODUCTCATEGORY
FROM (
  SELECT DISTINCT 
    PRODUCTID,
    PRODUCTNAME,
    PRODUCTCATEGORY
  FROM ORDERDWH.TRANS_SALES
) AS T;
```

```sql
INSERT INTO ORDERDWH.DIMPRODUCTS
SELECT * FROM ORDERDWH.VIEW_DIMPRODUCTS;
```

```sql
SELECT * FROM ORDERDWH.DIMPRODUCTS;
```

##### DIMREGION

```sql
CREATE OR REPLACE TABLE ORDERDWH.DIMREGION
(
  DIMREGIONKEY INT,
  REGIONID INT,
  REGIONNAME STRING,
  COUNTRY STRING
)
```

```sql
CREATE OR REPLACE VIEW ORDERDWH.VIEW_DIMREGION AS
SELECT 
  ROW_NUMBER() OVER (ORDER BY REGIONID) AS DIMREGIONKEY,
  REGIONID,
  REGIONNAME,
  COUNTRY
FROM (
  SELECT DISTINCT 
    REGIONID,
    REGIONNAME,
    COUNTRY
  FROM ORDERDWH.TRANS_SALES
) AS T;
```

```sql
INSERT INTO ORDERDWH.DIMREGION
SELECT * FROM ORDERDWH.VIEW_DIMREGION;
```

```sql
SELECT * FROM ORDERDWH.DIMREGION;
```

##### DIMDATE

```sql
CREATE TABLE ORDERDWH.DIMDATE
(
  DIMDATEKEY INT,
  ORDERID INT,
  ORDERDATE DATE
)
```

```sql
CREATE OR REPLACE VIEW ORDERDWH.VIEW_DIMDATE AS
SELECT 
  ROW_NUMBER() OVER (ORDER BY ORDERID) AS DIMORDERKEY,
  ORDERID,
  ORDERDATE
FROM (
  SELECT DISTINCT 
    ORDERID,
    ORDERDATE
  FROM ORDERDWH.TRANS_SALES
) AS T;
```

```sql
INSERT INTO ORDERDWH.DIMDATE
SELECT * FROM ORDERDWH.VIEW_DIMDATE;
```

```sql
SELECT * FROM ORDERDWH.DIMDATE;
```

```sql
CREATE OR REPLACE TABLE ORDERDWH.FACTSALES
(
  ORDERID INT,
  QUANTITY DECIMAL,
  UNITPRICE DECIMAL,
  TOTALAMOUNT DECIMAL,
  DIMPRODUCTKEY INT,
  DIMCUSTOMERKEY INT,
  DIMREGIONKEY INT,
  DIMDATEKEY INT
)
```

```sql
SELECT F.ORDERID, F.QUANTITY, F.UNITPRICE, F.TOTALAMOUNT, DC.DIMCUSTOMERKEY, DP.DIMPRODUCTKEY, DD.DIMDATEKEY, DR.DIMREGIONKEY
FROM ORDERDWH.TRANS_SALES F
LEFT JOIN ORDERDWH.DIMCUSTOMERS DC ON F.CUSTOMERID = DC.CUSTOMERID
LEFT JOIN ORDERDWH.DIMPRODUCTS DP ON F.PRODUCTID = DP.PRODUCTID
LEFT JOIN ORDERDWH.DIMDATE DD ON F.ORDERDATE = DD.ORDERDATE
LEFT JOIN ORDERDWH.DIMREGION DR ON F.COUNTRY = DR.COUNTRY;
```
